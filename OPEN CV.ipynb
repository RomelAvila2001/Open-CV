### Guardamos un archivo de video

```python
# importamos librerias 
import cv2 as cv 
import numpy as np 
# cremos un objeto VideoCapture 
cap = cv.VideoCapture("Curso.mp4") 
# obtenemos los fotogramas por segundo 
fps = cap.get(cv.CAP_PROP_FPS) 
# obtenemos el ancho y alto del video 
size = (int(cap.get(cv.CAP_PROP_FRAME_WIDTH)),int(cap.get(cv.CAP_PROP_FRAME_HEIGHT))) # definimos el codec 
fourcc = cv.VideoWriter_fourcc(*'XVID') 
# creamos un objeto VideoWrite 
v_Writer = cv.VideoWriter("Curso.avi", fourcc, 30, size) 
# creamos un bucle 
while(cap.isOpened()): 
    # leemos los cuadros del video 
    ret, frame = cap.read() 
    if(ret): 
        # convertimos la imagen a escala de grises 
        #frame = cv.cvtColor(frame, cv.COLOR_RGB2GRAY) 
        # giramos la imagen 
        frame = cv.flip(frame,0) 
        # guardamos el video 
        v_Writer.write(frame) 
        cv.imshow('frame', frame) 
        if(cv.waitKey(1)& 0xFF ==ord('q')): 
            break 
        else: 
            break 
v_Writer.release()  
cap.release() 
cv.destroyAllWindows() 
```

### Tomar Fotos

```python
#Importamos librerias
import cv2 as cv
import numpy as np

##Lectura del video
cap = cv.VideoCapture(0)
cont=1
while(cap.isOpened()):
    
    ret,frame = cap.read()
    
    if ret:

        cv.imshow('Image', frame)
        
        if cv.waitKey(1) & 0xFF==ord("a"):
            cv.imwrite("fotos/navidad_"+str(cont)+".jpg",frame)
            cont=cont+1
            
        elif cv.waitKey(10) & 0xFF==ord("q"):
            break
    else:
        break
cap.release()        
cv.destroyAllWindows()
```

### Filtrado de Imagenes

### Deteccion de bordes 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#img = cv.imread("one_1.jpg")
img = cv.imread("nami.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

sobel_x = cv.Sobel(img_gray,cv.CV_16U,1,0,ksize=5)
sobel_y = cv.Sobel(img_gray,cv.CV_16U,0,1,ksize=5)

sobel = sobel_x + sobel_y

plt.figure(1),plt.imshow(img_gray,cmap="gray"),plt.title("Original GRAY")
plt.xticks([]),plt.yticks([])
plt.figure(2),plt.imshow(sobel_x,cmap="gray"),plt.title("Sobel_X")
plt.xticks([]),plt.yticks([])
plt.figure(3),plt.imshow(sobel_y,cmap="gray"),plt.title("Sobel_Y")
plt.xticks([]),plt.yticks([])
plt.figure(4),plt.imshow(sobel,cmap="gray"),plt.title("Sobel")
plt.xticks([]),plt.yticks([])
plt.show()
```

### Template Maching

```python
# Creacion del template
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
img = cv.imread("messi.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
template = img[10:70,305:360]
cv.imwrite("template.jpg",template)

plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(template)
plt.show()
```

```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
# cargamos la Imagen
img = cv.imread("messi.jpg")
# cabiamos el espacio de color a escala de grises
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# cargamos el template
template = cv.imread("template.jpg")
# obtenemos la forma del template
w, h = template.shape[:2]
# listamos los 6 metodos de comparacion del template
metodos = ['cv.TM_CCOEFF', 'cv.TM_CCOEFF_NORMED', 'cv.TM_CCORR',
            'cv.TM_CCORR_NORMED', 'cv.TM_SQDIFF', 'cv.TM_SQDIFF_NORMED']

# ciclo para probar todos los metodos de comparacion del template
for meth in metodos:
   
    metodo = eval(meth)
    # Aplicamos template Matching
    res = cv.matchTemplate(img,template,metodo)
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    # SI el metodo es  TM_SQDIFF o TM_SQDIFF_NORMED, tomamos los minimos:
    if metodo in [cv.TM_SQDIFF, cv.TM_SQDIFF_NORMED]:
        top_left = min_loc
    # caso contrario tomamos el maximo
    else:
        top_left = max_loc
    # creamos la segunda cordenada para el rectangulo delimitador    
    bottom_right = (top_left[0] + w, top_left[1] + h)
    # graficamos el rectangulo en la imagen
    cv.rectangle(img,top_left, bottom_right, 255, 2)
    # Graficamos los resultados
    plt.subplot(121),plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.suptitle(meth)
    plt.show()
```
